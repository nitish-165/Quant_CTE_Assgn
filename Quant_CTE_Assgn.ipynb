```python
import pandas as pd
import numpy as np
import datetime as dt
import matplotlib.pyplot as plt
from matplotlib import style
import requests
style.use('ggplot')
plt.rcParams["figure.figsize"] = (20,10)
import warnings
warnings.filterwarnings('ignore')
```

```python
import talib as ta
```

```python
import yfinance as yf
```

```python
a1 = yf.Ticker("TCS.NS")
df1 = a1.history(start='2013-01-01', end='2018-12-31' , interval = '1d')
```

```python
df1
```

```python
df1['MACD'],df1['MACDSignal'],df1['MACDhist'] = ta.MACD(df1.Close,12,26,9)
df1['RSI'] = ta.RSI(df1.Close,14)
df1['EMA50'] = ta.EMA(df1.Close,50)
df1['EMA200'] = ta.EMA(df1.Close,200)
```

```python
df1
```

```python
df1_new=df1.copy()
df1_new=df1_new.drop(df1_new.index[0:199])
df1_new
```

```python
position=0
position_index=0
target= 10
stoploss= 2
realised_pnl=0
number_of_trades=0
pnl_l = 0
pnl_h = 0
pnl_l1 = []
position_l = []
tradebook = {}

for i in range(len(df1_new)-1):
    if ((position == 0)&(df1_new.MACD.loc[df1_new.index[i]]<df1_new.MACDSignal.loc[df1_new.index[i]])
         & (df1_new.MACDhist.loc[df1_new.index[i]]<2)):
            position_time=df1_new.index[i].time()
            entry_price = df1_new.loc[df1_new.index[i+1],'Open']
            position_index=df1_new.index[i]
            position = 1
            print('long position at {}'.format(df1_new.index[i]))
            number_of_trades=number_of_trades+1
            position_l.append(position_index)
            continue
    
    if ((position == 0)&(df1_new.MACD.loc[df1_new.index[i]]<df1_new.MACDSignal.loc[df1_new.index[i]])
         & (df1_new.MACDhist.loc[df1_new.index[i]]<5))& (df1_new.RSI.loc[df1_new.index[i]]<30):
            position_time=df1_new.index[i].time()
            entry_price = df1_new.loc[df1_new.index[i+1],'Open']
            position_index=df1_new.index[i]
            position = 1
            print('long position at {}'.format(df1_new.index[i]))
            number_of_trades=number_of_trades+1
            position_l.append(position_index)
            continue
            
    if ((position == 0)&(df1_new.MACD.loc[df1_new.index[i]]<df1_new.MACDSignal.loc[df1_new.index[i]])
         & (df1_new.MACDhist.loc[df1_new.index[i]]<5)
         & (df1_new.EMA50.loc[df1_new.index[i]]>df1_new.EMA200.loc[df1_new.index[i]])):
            position_time=df1_new.index[i].time()
            entry_price = df1_new.loc[df1_new.index[i+1],'Open']
            position_index=df1_new.index[i]
            position = 1
            print('long position at {}'.format(df1_new.index[i]))
            number_of_trades=number_of_trades+1
            position_l.append(position_index)
            continue
    
        
    if ((position == 0)&(df1_new.MACD.loc[df1_new.index[i]]>df1_new.MACDSignal.loc[df1_new.index[i]])
         &(df1_new.MACDhist.loc[df1_new.index[i]]<2)):
            position_time=df1_new.index[i].time()
            position_index=df1_new.index[i]
            entry_price = df1_new.loc[df1_new.index[i],'Open']
            position = -1
            print('short position at {}'.format(df1_new.index[i]))
            number_of_trades=number_of_trades+1
            position_l.append(position_index)
            continue
    
    if ((position == 0)&(df1_new.MACD.loc[df1_new.index[i]]>df1_new.MACDSignal.loc[df1_new.index[i]])
         &(df1_new.MACDhist.loc[df1_new.index[i]]<5)& (df1_new.RSI.loc[df1_new.index[i]]>70)):
            position_time=df1_new.index[i].time()
            position_index=df1_new.index[i]
            entry_price = df1_new.loc[df1_new.index[i],'Open']
            position = -1
            print('short position at {}'.format(df1_new.index[i]))
            number_of_trades=number_of_trades+1
            position_l.append(position_index)
            continue
            
    if ((position == 0)&(df1_new.MACD.loc[df1_new.index[i]]>df1_new.MACDSignal.loc[df1_new.index[i]])
         &(df1_new.MACDhist.loc[df1_new.index[i]]<5)
         &(df1_new.EMA50.loc[df1_new.index[i]]<df1_new.EMA200.loc[df1_new.index[i]])):
            position_time=df1_new.index[i].time()
            position_index=df1_new.index[i]
            entry_price = df1_new.loc[df1_new.index[i],'Open']
            position = -1
            print('short position at {}'.format(df1_new.index[i]))
            number_of_trades=number_of_trades+1
            position_l.append(position_index)
            continue
    
    if (position_index!=0):
            pnl = ((df1_new.loc[df1_new.index[i],'Close']- entry_price)/entry_price)*(position)*100  
    
    if ((position != 0) and ((pnl  >target) or (pnl < -stoploss))):
        realised_pnl=realised_pnl+pnl
        pnl_l1.append(pnl)
        position=0
        print('position exited at {}'.format(df1_new.index[i].time()))
        position_index=0
        tradebook[number_of_trades] = {'entry_time' :position_time ,'entry_price': entry_price ,'long_short': position ,'exit_time' :df1_new.index[i], 'pnl' :pnl}
        print(pnl)
  
print(realised_pnl)
        
        
        
       
```

```python
tradebook1 = pd.DataFrame(tradebook).T
tradebook1['Profit'] = (tradebook1['entry_price']*tradebook1['pnl'])/100
```

```python
number_of_trades
```

```python
plt.plot()
plt.plot(np.cumsum(pnl_l1))
```
